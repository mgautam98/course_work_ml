```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd
import torch
from torch.autograd import Variable
import torch.nn.functional as F
import cv2
```

```python
img = cv2.imread("background2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(0)
plt.imshow(img, cmap="gray")
```

```python
kernel = np.array([[-1, -1, -1],
                   [0, 0, 0], 
                   [1, 1, 1]])
```

```python
new_img = np.zeros((img.shape[0]-kernel.shape[0], img.shape[1]-kernel.shape[1]))
```

```python
for ix in range(new_img.shape[0]):
    for iy in range(new_img.shape[1]):
        img_patch = img[ix:ix+kernel.shape[0], iy:iy+kernel.shape[1]]
        
        h_prod = img_patch*kernel
        new_img[ix, iy] = max(0, h_prod.sum())
```

```python
plt.imshow(new_img.astype(np.uint8), cmap='gray')
```
